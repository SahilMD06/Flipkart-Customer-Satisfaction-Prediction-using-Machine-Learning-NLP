## 3. ***Data Wrangling***

### Data Wrangling Code

````python

# Data Wrangling - Making Dataset Analysis Ready

# Create Copy of Dataset
df1 = df.copy()

# Remove Duplicate Values
df1.drop_duplicates(inplace=True)

# Handling Missing Values

# Numerical Columns -> Fill with Median
num_cols = df1.select_dtypes(include=np.number).columns

for col in num_cols:
    df1[col].fillna(df1[col].median(), inplace=True)

# Categorical Columns -> Fill with Mode
cat_cols = df1.select_dtypes(include='object').columns

for col in cat_cols:
    df1[col].fillna(df1[col].mode()[0], inplace=True)

# Convert Date Columns into Datetime Format
date_cols = ['order_date_time',
             'Issue_reported at',
             'issue_responded',
             'Survey_response_Date']

for col in date_cols:
    df1[col] = pd.to_datetime(df1[col], errors='coerce')

# Extract Useful Features from Date Columns
df1['Order_Year'] = df1['order_date_time'].dt.year
df1['Order_Month'] = df1['order_date_time'].dt.month
df1['Order_Day'] = df1['order_date_time'].dt.day

# Encoding Categorical Variables
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = df1.select_dtypes(include='object').columns

for col in cat_cols:
    df1[col] = le.fit_transform(df1[col].astype(str))

# Final Dataset Information
print("Dataset Shape :", df1.shape)

print("\nMissing Values After Cleaning:\n")
print(df1.isnull().sum())

print("\nFirst 5 Rows of Cleaned Dataset:\n")
df1.head()
````

### What all manipulations have you done and insights you found?

Several data preprocessing and manipulation steps were performed on the dataset to improve data quality and prepare it for analysis and machine learning.

Missing values and duplicate records were identified and handled appropriately. Date and time columns were converted into proper datetime format for easier analysis. Categorical variables such as channel_name, Product_category, Agent Shift, and Tenure Bucket were encoded for machine learning purposes.

Outliers in numerical columns like connected_handling_time and Item_price were analyzed. Feature engineering techniques were also applied to derive meaningful insights from date and time columns.

The analysis revealed that handling time, product category, and agent experience significantly influence customer satisfaction scores. Certain product categories generated higher complaint volumes, while experienced agents achieved better CSAT ratings compared to less experienced agents.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀)

There is no significant relationship between connected handling time and CSAT Score. The amount of time taken to resolve customer issues does not affect customer satisfaction levels.

Alternate Hypothesis (H₁)

There is a significant relationship between connected handling time and CSAT Score. Customers are more likely to provide lower satisfaction scores when the issue handling time increases.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Hypothesis Testing 1

from scipy.stats import pearsonr

# Selecting Variables
x = df1['connected_handling_time']
y = df1['CSAT Score']

# Pearson Correlation Test
corr, p_value = pearsonr(x, y)

print("Correlation Coefficient :", corr)
print("P-Value :", p_value)

# Conclusion
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test :- The Pearson Correlation Test was performed to obtain the P-value. This statistical test measures the strength and direction of the linear relationship between two numerical variables. In this project, connected handling time and CSAT Score are numerical variables, making Pearson Correlation an appropriate statistical method.

##### Why did you choose the specific statistical test?

Pearson Correlation Test was selected because the objective was to determine whether a relationship exists between customer issue handling time and customer satisfaction score. Both variables are continuous numerical variables, and the test helps identify whether longer handling time negatively affects customer satisfaction. This insight is highly important for improving customer support efficiency and enhancing user experience on the Flipkart platform.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀)

Different agent shifts do not have any significant impact on connected handling time. All shifts perform similarly in resolving customer issues.

Alternate Hypothesis (H₁)

Different agent shifts significantly impact connected handling time. Certain shifts may take more time to resolve customer issues compared to others.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Hypothesis Testing 2

from scipy.stats import f_oneway

# Creating Groups
morning = df1[df1['Agent Shift'] == 'Morning']['connected_handling_time']
evening = df1[df1['Agent Shift'] == 'Evening']['connected_handling_time']
night = df1[df1['Agent Shift'] == 'Night']['connected_handling_time']

# ANOVA Test
f_stat, p_value = f_oneway(morning, evening, night)

print("F-Statistic :", f_stat)
print("P-Value :", p_value)

# Conclusion
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

ANOVA (Analysis of Variance) Test :- The ANOVA (Analysis of Variance) Test was performed to obtain the P-value. ANOVA is used when comparing the means of more than two groups to determine whether significant differences exist among them.

##### Why did you choose the specific statistical test?

ANOVA Test was chosen because the dataset contains multiple categories of agent shifts such as Morning, Evening, and Night shifts. The goal was to compare the average connected handling time across these different groups. This test helps identify whether workload distribution or staffing differences during specific shifts affect customer service performance. Understanding these differences can help Flipkart optimize staffing and improve operational efficiency.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀)

Product category and CSAT Score are independent of each other. Different product categories do not influence customer satisfaction levels.

Alternate Hypothesis (H₁)

Product category and CSAT Score are associated with each other. Certain product categories receive significantly different customer satisfaction scores.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Hypothesis Testing 3

from scipy.stats import chi2_contingency

# Contingency Table
table = pd.crosstab(df1['Product_category'],
                    df1['CSAT Score'])

# Chi-Square Test
chi_stat, p_value, dof, expected = chi2_contingency(table)

print("Chi-Square Statistic :", chi_stat)
print("P-Value :", p_value)

# Conclusion
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test :- The Chi-Square Test of Independence was performed to obtain the P-value. This test determines whether there is a statistically significant association between two categorical variables.

##### Why did you choose the specific statistical test?

The Chi-Square Test was selected because both Product Category and CSAT Score are categorical variables in the dataset. The objective was to determine whether customer satisfaction differs across various product categories. This analysis is important because it helps identify problematic product segments that may require quality improvements, better customer support, or enhanced delivery services. ## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

# Checking Missing Values
df1.isnull().sum()

# Filling Numerical Missing Values with Median
num_cols = df1.select_dtypes(include=np.number).columns

for col in num_cols:
    df1[col].fillna(df1[col].median(), inplace=True)

# Filling Categorical Missing Values with Mode
cat_cols = df1.select_dtypes(include='object').columns

for col in cat_cols:
    df1[col].fillna(df1[col].mode()[0], inplace=True)

# Verifying Missing Values
df1.isnull().sum()
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

In this project, both median imputation and mode imputation techniques were used to handle missing values. For numerical columns, missing values were replaced using the median because median is less sensitive to outliers and extreme values. This helps maintain the original distribution of numerical features such as connected handling time and item price.

For categorical columns, missing values were replaced using the mode, which represents the most frequently occurring category in a column. Since categorical variables like product category, agent shift, and communication channel contain repeated categories, mode imputation helps preserve the most common business patterns in the dataset. Proper handling of missing values improves data quality and ensures better model performance during machine learning.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier Treatments

# Boxplot Before Treatment
plt.figure(figsize=(10,5))
sns.boxplot(x=df1['connected_handling_time'])
plt.title("Boxplot Before Outlier Treatment")
plt.show()

# IQR Method
Q1 = df1['connected_handling_time'].quantile(0.25)
Q3 = df1['connected_handling_time'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Removing Outliers
df1 = df1[(df1['connected_handling_time'] >= lower_limit) &
          (df1['connected_handling_time'] <= upper_limit)]

# Boxplot After Treatment
plt.figure(figsize=(10,5))
sns.boxplot(x=df1['connected_handling_time'])
plt.title("Boxplot After Outlier Treatment")
plt.show()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The IQR (Interquartile Range) method was used to detect and handle outliers in the dataset. This technique identifies extreme values that fall significantly below or above the normal data range. Outliers were mainly present in numerical columns such as connected handling time and item price.

The IQR method was selected because it is highly effective for skewed datasets and does not assume normal distribution of data. Removing extreme outlier values helps improve model stability, reduce noise, and prevent biased predictions during machine learning model training. ##### Do you think the dataset is imbalanced? Explain Why.

Yes, the dataset may contain imbalance in the target variable, especially in CSAT Score categories. Certain satisfaction scores may occur much more frequently than others, causing the machine learning model to become biased toward majority classes.

An imbalanced dataset can reduce prediction accuracy for minority classes and negatively affect model performance. Therefore, handling class imbalance is important to ensure fair and accurate prediction across all customer satisfaction categories.

````python
# Handling Imbalanced Dataset using SMOTE

from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Create Copy
X_smote = X_train.copy()

# Convert Datetime Columns to Integer
for col in X_smote.select_dtypes(include='datetime64[ns]').columns:
    X_smote[col] = X_smote[col].astype('int64')

# Encode Object Columns
le = LabelEncoder()

for col in X_smote.select_dtypes(include='object').columns:
    X_smote[col] = le.fit_transform(
        X_smote[col].astype(str)
    )

# Fill Missing Values
X_smote = X_smote.fillna(0)

# Apply SMOTE
smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(
    X_smote,
    y_train
)

# Display Shapes
print("Original Shape :", X_train.shape)

print("Resampled Shape :", X_resampled.shape)

print("Original Target Shape :", y_train.shape)

print("Resampled Target Shape :", y_resampled.shape)
````

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

SMOTE (Synthetic Minority Oversampling Technique) was used to handle the imbalanced dataset. This technique creates synthetic data points for minority classes instead of simply duplicating existing records. By generating new synthetic samples, SMOTE helps balance the target variable distribution and improves the learning capability of the machine learning model.

SMOTE was selected because the CSAT Score categories in the dataset were imbalanced, where some satisfaction classes appeared much more frequently than others. Without balancing, the model may become biased toward majority classes and perform poorly on minority classes. Applying SMOTE improves model fairness, prediction accuracy, and overall classification performance across all customer satisfaction categories.

F1-score provides a balanced measure between precision and recall. A higher F1-score indicates that the model performs consistently well across all classes. Overall, the machine learning model helps businesses improve customer support quality, identify critical customer issues faster, enhance customer satisfaction, and support better business decision-making.

### ML Model - 3

````python
# ML Model - 3 Implementation

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

# Initialize Logistic Regression Model
lr_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

# Train Model
lr_model.fit(X_train_ml, y_train)

# Prediction
y_pred_lr = lr_model.predict(X_test_ml)

# Accuracy
lr_accuracy = accuracy_score(
    y_test,
    y_pred_lr
)

print("Accuracy Score :",
      lr_accuracy)

# Classification Report
print("\nClassification Report:\n")

print(classification_report(
    y_test,
    y_pred_lr
))

# Confusion Matrix
print("\nConfusion Matrix:\n")

print(confusion_matrix(
    y_test,
    y_pred_lr
))
````

Logistic Regression was used as the third machine learning model for predicting customer satisfaction scores. Logistic Regression is a supervised learning classification algorithm that predicts the probability of different target classes based on input features. It is simple, efficient, and performs well on structured datasets with properly scaled features.

The model performance was evaluated using accuracy score, precision, recall, F1-score, and confusion matrix.
````python
# Visualizing evaluation Metric Score chart


from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score
)

import matplotlib.pyplot as plt

# Evaluation Metrics
accuracy = accuracy_score(
    y_test,
    y_pred_lr
)

precision = precision_score(
    y_test,
    y_pred_lr,
    average='weighted'
)

recall = recall_score(
    y_test,
    y_pred_lr,
    average='weighted'
)

f1 = f1_score(
    y_test,
    y_pred_lr,
    average='weighted'
)

# Create Dictionary
scores = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
}

# Plot Chart
plt.figure(figsize=(8,5))

plt.bar(scores.keys(),
        scores.values())

plt.title("Logistic Regression Evaluation Metrics")

plt.ylabel("Score")

plt.ylim(0,1)

plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# Hyperparameter Tuning using GridSearchCV

from sklearn.model_selection import GridSearchCV

# Parameter Grid
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear']
}

# GridSearchCV
grid_search_lr = GridSearchCV(
    estimator=LogisticRegression(
        max_iter=1000,
        random_state=42
    ),
    param_grid=param_grid,
    cv=3,
    n_jobs=1,
    verbose=2
)

# Train GridSearch
grid_search_lr.fit(
    X_train_ml,
    y_train
)

# Best Parameters
print("Best Parameters :")

print(grid_search_lr.best_params_)

# Best Model
best_lr_model = (
    grid_search_lr.best_estimator_
)

# Prediction
y_pred_best_lr = best_lr_model.predict(
    X_test_ml
)

# Accuracy
best_lr_accuracy = accuracy_score(
    y_test,
    y_pred_best_lr
)

print("\nImproved Accuracy :",
      best_lr_accuracy)
````
